## Comentarios


```python
data1 = pd.read_csv("Dataset_benigno_ransomware/Dataset01.csv")
```

```python
def definir( label_v ):
    if(label_v == "Benigno"):
        return 1
    else:
        return 0
```

```python
data1["Definidos"] = data1["Label"].apply(definir)
```

```python
data1
```

```python
data1.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_Y = data1["Definidos"].copy()
```

```python
data_Y
```

```python
data_X = data1.drop(["Definidos","Label"], axis = 1).copy()
```

```python
data_X
```

```python
train_ratio = 0.7 
test_ratio = 0.2 
validation_ratio = 0.1
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, test_size=test_ratio, random_state=101)
```

```python
X_train
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=validation_ratio/(validation_ratio + train_ratio), random_state=101)
```

```python
X_train
```

```python
X_train.shape[1]
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu', input_shape = (X_train.shape[1],)))
```

```python
model.add(tf.keras.layers.Dropout(0.2))
```

```python
model.add(tf.keras.layers.Dense(units = 2, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['sparse_categorical_accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs =10, validation_data=(X_val, y_val))
```

##Testeo ahora :D 

```python
test_lossANN, test_accuracyANN = model.evaluate(X_test, y_test)
```
