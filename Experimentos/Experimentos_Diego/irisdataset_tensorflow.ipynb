## Iris data set con TensorFlow


```python
#Importando librerias 

import tensorflow as tf
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Cargando dataset

path_file = 'C:\\Users\\dieku\\Documents\\GitHub\\cyberseguridad-code\\Experimentos\\Experimentos_Diego\\Datasets\\iris.csv'

data = pd.read_csv(path_file)

data
```

```python
data['setosa'].unique()
```

```python
def define_tipo(x):
    if(x == 'setosa'):
        return 0
    elif(x == 'virginica'):
        return 1
    else:
        return 2

data_mod = data.copy()
data_mod['clase'] = data['setosa'].apply(define_tipo)
data_mod
```

```python
data_mod.corr()
```

```python
dataX = data_mod.drop(['setosa', 'clase'], axis=1).copy()
dataY = data_mod[['clase']].copy()
```

```python
#Correlacion de variables.
dataX.corr()
```

```python
dataY
```

```python
dataX = (dataX-np.mean(dataX, axis=0)) / np.std(dataX,axis=0)
dataY = (dataY-np.mean(dataY, axis=0)) / np.std(dataY,axis=0)
```

### Generacion de datos de entrenamiento, testing y validacion o.o

 E, T, V = 70%, 20%, 10%

```python
from sklearn.model_selection import train_test_split

train_ratio = 0.7
test_ratio = 0.2
validation_ratio = 0.1
```

```python

X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=test_ratio, random_state=101)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=validation_ratio/(validation_ratio + train_ratio), random_state=101)
```

```python
print("===============================================================")
print("Porcentaje de entrenamiento X: "+ str(X_train.shape[0]/dataX.shape[0]) +"\n")
print("Porcentaje de testeo X: "+ str(X_test.shape[0]/dataX.shape[0]) +"\n")
print("Porcentaje de validacion X: "+ str(X_val.shape[0]/dataX.shape[0]) +"\n")
print("===============================================================")
print("Porcentaje de entrenamiento Y: "+ str(y_train.shape[0]/dataY.shape[0]) +"\n")
print("Porcentaje de testeo Y: "+ str(y_test.shape[0]/dataY.shape[0]) +"\n")
print("Porcentaje de validacion Y: "+ str(y_val.shape[0]/dataY.shape[0]) +"\n")
print("===============================================================")
```

```python
modelANN = tf.keras.models.Sequential()
```

```python
modelANN.add(tf.keras.layers.Dense(units=5, activation='relu',input_shape=(X_train.shape[1],)))
modelANN.add(tf.keras.layers.Dropout(0.2))
modelANN.add(tf.keras.layers.Dense(units=3,activation='softmax'))
```

```python
modelANN.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
modelANN.summary()
```

```python
historyANN = modelANN.fit(X_train,y_train, epochs=5, validation_data=(X_val,y_val['clase']))
```

```python
test_loss, test_accuracy = modelANN.evaluate(X_test, y_test)
```
